```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(10)
data = pd.DataFrame({
    'Value': np.random.rand(100),
    'Group': np.random.choice(['A', 'B', 'C'], 100),
    'Category': np.random.choice(['X', 'Y', 'Z'], 100),
    'Numeric': np.random.randint(1, 20, 100)
})
print(data.head())

```

```python
# Enhanced Pairplot
sns.pairplot(data, hue='Group', aspect=1.5)
plt.suptitle('Pairplot with Hue by Group', verticalalignment='baseline', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Complex Heatmap
corr = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Complex Heatmap of Correlation')
plt.show()
```

```python
# Advanced Boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(x='Group', y='Numeric', hue='Category', data=data, palette="coolwarm")
plt.title('Advanced Boxplot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
import pandas as pd

# Sample data snippet as a pandas DataFrame
data = {
    'Value': [0.771321, 0.020752, 0.633648, 0.748804, 0.498507],
    'Group': ['A', 'A', 'B', 'B', 'A'],
    'Category': ['Z', 'Z', 'Y', 'Y', 'X'],
    'Numeric': [10, 17, 18, 11, 17]
}
df = pd.DataFrame(data)

# Perform the Mann-Whitney U test on the 'Value' column for groups 'A' and 'B'
group_A_values = df[df['Group'] == 'A']['Value']
group_B_values = df[df['Group'] == 'B']['Value']
stat, p = mannwhitneyu(group_A_values, group_B_values)

# Create a boxplot
plt.figure(figsize=(10, 8))
box_plot = sns.boxplot(x='Group', y='Value', data=df)

# Annotate with the p-value
x_position = 0.5
max_value = df['Value'].max()
y_position = max_value + (max_value * 0.1)  # 10% above the max value for clearance
plt.text(x_position, y_position, f'Mann-Whitney U test p-value: {p:.3f}',
         horizontalalignment='center', fontsize=12)

plt.title('Value Distribution by Group with Mann-Whitney U test p-value')
plt.show()
```
